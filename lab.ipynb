```python
impute=SimpleImputer(strategy='most_frequent')
new_df=impute.fit_transform(df)
df.columns
```

```python
x=pd.DataFrame(new_df)
cols=pd.Series(['region', 'province', 'name', 'tc', 'cres', 'csor', 'istat', 'area',
       'pop', 'alt', 'isle', 'sea', 'pden', 'wden', 'urb', 'fee', 'd_fee',
       'sample', 'organic', 'paper', 'glass', 'wood', 'metal', 'plastic',
       'raee', 'texile', 'other', 'msw_so', 'msw_un', 'msw', 'sor', 'geo',
       'roads', 's_wteregio', 's_landfill', 'gdp', 'proads', 'wage',
       'finance'])
```

```python
x.rename(columns=cols,inplace=True)
```

```python
x.isnull().sum()
```

```python
x.drop(columns=['region','province','name'],inplace=True)
x
```

```python
X = np.array(x)
X = X[:,0:-2]
y = X[:,-1]
```
